```python
import requests
from bs4 import BeautifulSoup
import time

def get_stock_price(url, class_name):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    price = soup.find(class_=class_name).text.strip()
    return price

# Define URLs for NSE and BSE
urls = {
    'INFY': {
        'url': 'https://www.google.com/finance/quote/INFY:NSE',
        'exchange': 'NSE'
    },
    '500209': {
        'url': 'https://www.google.com/finance/quote/500209:BOM',
        'exchange': 'BSE'
    }
}

# Define class names for extracting price
class_names = {
    'INFY': 'YMlKec fxKbKc',
    '500209': 'YMlKec fxKbKc'
}

while True:
    for ticker, data in urls.items():
        url = data['url']
        exchange = data['exchange']
        class_name = class_names[ticker]

        try:
            price = get_stock_price(url, class_name)
            print(f'{ticker} ({exchange}): {price}')
        except Exception as e:
            print(f'Error retrieving data for {ticker} ({exchange}): {str(e)}')

    time.sleep(10)  # Wait for 10 seconds before fetching data again
```

```python
# import libraries
import pandas as pd
import numpy as np
from sklearn import metrics
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("infy_22.csv")
df.head()
```

```python
series_shifted=df['Close Price'].shift()
df['prev_closeprice']= series_shifted
df.head()
```

```python
#time format
df['Date']=pd.to_datetime(df.Date)
df.head()
```

```python
df.shape
```

```python
#null entries
df.isnull().sum()
```

```python
#drop/remove
df.dropna(inplace=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
#plot close prices
df['Close Price'].plot(figsize=(10,6))
```

```python
#x and y
x=df[['Open Price','prev_closeprice']]
y=df['Close Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#print regressor
print(regressor.coef_)
print(regressor.intercept_)
```

```python
#predicted value
y_pred=regressor.predict(x_test)
print(x_test)
print(y_pred)
```

```python
import math
print("Mean Absolute Error", metrics.mean_absolute_error(y_test,y_pred))
print("Mean Square Error", metrics.mean_squared_error(y_test,y_pred))
print("Root Mean Error", math.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
